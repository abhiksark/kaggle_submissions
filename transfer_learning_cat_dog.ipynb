```
train_dir = './train/'
test_dir = './test1/'
```

```
from keras.applications import VGG16
 
conv_base = VGG16(weights='imagenet',
                  include_top=False,
                  input_shape=(224, 224, 3))
```

```
nTrain = 15000
nImages = 19000
nVal = 4000

train_features = np.zeros(shape=(nTrain, 7, 7, 512))
train_labels = np.zeros(shape=(nTrain,1))

```

```
datagen = ImageDataGenerator(rescale=1./255)
batch_size = 20
 
train_features = np.zeros(shape=(nTrain, 7, 7, 512))
train_labels = np.zeros(shape=(nTrain))
 
train_generator = datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='binary')
```

```
datagen = ImageDataGenerator(rescale=1./255)
batch_size = 20
 
val_features = np.zeros(shape=(nVal, 7, 7, 512))
val_labels = np.zeros(shape=(nVal))
 
val_generator = datagen.flow_from_directory(
    './val/',
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='binary')
```

```
i = 0
nSize = 0
for inputs_batch, labels_batch in train_generator:
    features_batch = conv_base.predict(inputs_batch)
    if features_batch.shape[0]:   
      nSize = features_batch.shape[0]+nSize
      #print(nSize)
      try:
        train_features[i * batch_size : (i + 1) * batch_size] = features_batch
        train_labels[i * batch_size : (i + 1) * batch_size] = labels_batch
      except:
        break
    if(nSize>=nTrain):
      break

         
len(train_features)
train_features = np.reshape(train_features, (nSize, 7 * 7 * 512))
len(train_features)
```

```
i = 0
nSize = 0
for inputs_batch, labels_batch in val_generator:
    features_batch = conv_base.predict(inputs_batch)
    if features_batch.shape[0]:   
      nSize = features_batch.shape[0]+nSize
      #print(nSize)
      try:
        val_features[i * batch_size : (i + 1) * batch_size] = features_batch
        val_labels[i * batch_size : (i + 1) * batch_size] = labels_batch
      except:
        break
    if(nSize>=nVal):
      break

         
val_features = np.reshape(val_features, (nSize, 7 * 7 * 512))
```

```
from keras import models
from keras import layers
from keras import optimizers
 
model = models.Sequential()
model.add(layers.Dense(256, activation='relu', input_dim=7 * 7 * 512))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(optimizer=optimizers.RMSprop(lr=2e-4),
              loss='binary_crossentropy',
              metrics=['acc'])
```

```
len(train_labels)
```

```

 
history = model.fit(train_features,
                    train_labels,
                    epochs=20,
                    batch_size=batch_size,
                    validation_data=(val_features,val_labels))
```

```
def prepare_data(list_of_images):
    """
    Returns two arrays: 
        x is an array of resized images
        y is an array of labels
    """
    x = [] # images as arrays
    y = [] # labels
    img_width = 224
    img_height = 224
    
    for image in list_of_images:
        #print(image)
        try:
          x.append(cv2.resize(cv2.imread(image), (img_width,img_height), interpolation=cv2.INTER_CUBIC))
        except:
          pass
  
  
    for i in list_of_images:
        if 'dog' in i:
            y.append(1)
        elif 'cat' in i:
            y.append(0)
        #else:
            #print('neither cat nor dog name present in images')
            
    return x, y
```

```
test_images_dogs_cats = [test_dir+i for i in os.listdir(test_dir)]
```

```
x=0
prediction_probabilities = []
name = []
i=0
for image in test_images_dogs_cats:
  img_width = 224
  img_height = 224
  x=(cv2.resize(cv2.imread(image), (img_width,img_height), interpolation=cv2.INTER_CUBIC))
  print(x[0].shape)
  x = np.array(x,dtype='float32')
  
  x = x * 1./255
  x= np.reshape(x,(1,224, 224, 3))
  features = conv_base.predict(x)
  features  = np.reshape(features,(1, 7* 7 * 512))
  
  name.append(int((image.split("/")[2]).split(".")[0]))
  prediction_probabilities.append(model.predict(features, verbose=1))
  print(i)
  i=i+1
```
